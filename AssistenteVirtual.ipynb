```python
from gtts import gTTS
import IPython.display as ipd

def falar(texto):
    tts = gTTS(texto, lang="pt")
    tts.save("fala.mp3")
    return ipd.Audio("fala.mp3")

# Teste
falar("Olá, sou seu assistente virtual!")

```

```python
import speech_recognition as sr

def ouvir_microfone():
    rec = sr.Recognizer()
    with sr.Microphone() as source:
        print("Diga algo...")
        audio = rec.listen(source)
    try:
        comando = rec.recognize_google(audio, language="pt-BR")
        print("Você disse:", comando)
        return comando.lower()
    except:
        print("Não entendi, pode repetir?")
        return ""
```

```python
import wikipedia
import webbrowser

def executar_comando(comando):
    if "youtube" in comando:
        falar("Abrindo o YouTube")
        webbrowser.open("https://youtube.com")

    elif "wikipedia" in comando:
        falar("Buscando no Wikipedia")
        resultado = wikipedia.summary(comando.replace("wikipedia", ""), sentences=2, auto_suggest=False)
        print(resultado)

    elif "farmácia" in comando:
        falar("Buscando farmácias próximas")
        webbrowser.open("https://www.google.com/maps/search/farmácia+perto+de+mim")

    else:
        falar("Desculpe, ainda não sei fazer isso.")

```

```python
import speech_recognition as sr
from google.colab import files

def ouvir_arquivo():
    # Upload de arquivo de áudio
    uploaded = files.upload()
    audio_path = list(uploaded.keys())[0]

    recognizer = sr.Recognizer()
    with sr.AudioFile(audio_path) as source:
        audio = recognizer.record(source)  # pega todo o áudio
    try:
        texto = recognizer.recognize_google(audio, language="pt-BR")
        return texto
    except:
        return "Não entendi o áudio."

# Teste
print("Fale algo em um arquivo de áudio e envie:")
texto = ouvir_arquivo()
print("Você disse:", texto)

```

```python
def executar_comando(comando):
    comando = comando.lower()

    if "youtube" in comando:
        falar("Abrindo YouTube...")
        webbrowser.open("https://www.youtube.com")

    elif "wikipedia" in comando:
        try:
            # remove a palavra "abrir wikipedia" e pega só o termo
            termo = comando.replace("abrir wikipedia", "").replace("wikipedia", "").strip()
            if termo == "":
                falar("Você não disse o que pesquisar na Wikipedia.")
            else:
                resumo = wikipedia.summary(termo, sentences=2, auto_suggest=False)
                falar(resumo)
        except Exception as e:
            falar("Não consegui encontrar esse termo na Wikipedia.")
            print("Erro:", e)

    elif "farmácia" in comando:
        falar("Buscando farmácias próximas")
        webbrowser.open("https://www.google.com/maps/search/farmácia+perto+de+mim")

    elif "sair" in comando:
        falar("Encerrando, até logo!")
        return False

    else:
        falar("Desculpe, ainda não sei fazer isso.")

    return True

# Lista de comandos para teste
comandos_teste = [
    "abrir youtube",
    "abrir wikipedia python",
    "abrir farmácia",
    "sair"
]

# Rodar cada comando em sequência
for comando in comandos_teste:
    print("\nVocê disse:", comando)
    executar_comando(comando)


```
